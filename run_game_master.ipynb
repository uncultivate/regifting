```python
from gift_strategies_master import *
from regifting import *
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

## Run Game

### Contestants
1. The Grinch 
2. Quack Quack Quack 
3. revelrous ryan 
4. Johann Carl Friedrich Gauss #1 Fan Club 
5. NoGift4U 
6. Avaricious Rapacity 
7. Harpo 
8. Grateful 

```python
# Initialize the game with gifter classes and number of presents
game = GiftingGame(
    gifter_classes = [
        TheGrinch, 
        NoGift4U, 
        RevelrousRyan, 
        QuackQuackQuack,
        MrGauss,
        GimmeGimmeGimme,
        Harpo,
        M_gifter
    ], 
    num_presents=100)

# Run the full tournament
final_results = game.run_tournament()
```

## Final Tally

```python
# Create a pandas DataFrame from the results
df_results = pd.DataFrame(list(final_results.items()), columns=['Gifter Type', 'Total Presents'])

# Sort the DataFrame by Total presents in descending order
df_results_sorted = df_results.sort_values('Total Presents', ascending=False).reset_index(drop=True)
```

```python

# Create color scale
# Create a list of 8 contrasting colors using hex codes
colors = [
    '#FF6B6B',  # Coral red
    '#4ECDC4',  # Turquoise
    '#45B7D1',  # Sky blue  
    '#96CEB4',  # Sage green
    '#FFEEAD',  # Pale yellow
    '#D4A5A5',  # Dusty rose
    '#9B59B6',  # Purple
    '#F1C40F'   # Golden yellow
]

colors = ['#FFA07A', '#98FB98', '#87CEFA', '#DDA0DD', '#F0E68C']

# Create the figure with subplots
fig = make_subplots(rows=1, cols=1)

# Add bars
fig.add_trace(go.Bar(
    y=df_results_sorted['Gifter Type'],
    x=df_results_sorted['Total Presents'],
    orientation='h',
    marker=dict(color=colors[:len(df_results_sorted)]),
    text=df_results_sorted['Total Presents'],
    textposition='outside',
    insidetextanchor='middle',
    textfont=dict(size=14),
    hoverinfo='text',
    hovertext=[f"{gifter}: {presents} presents" for gifter, presents in zip(df_results_sorted['Gifter Type'], df_results_sorted['Total Presents'])],
))

# Update layout
fig.update_layout(
    title={
        'text': "Gifter Leaderboard",
        'y':0.95,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top',
        'font': dict(size=24, color='#444444')
    },
    xaxis_title="Total Presents",
    font=dict(family="Arial", size=14, color="#444444"),
    plot_bgcolor='rgba(0,0,0,0)',
    xaxis=dict(showgrid=True, gridcolor='#DDDDDD', gridwidth=0.5),
    yaxis=dict(showgrid=False),
    margin=dict(l=100, r=20, t=70, b=50),
    height=500,
    width=800,
)

# Show the plot
fig.show()
```
